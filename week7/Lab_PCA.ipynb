# Lab: PCA

### Create Data

```python
#dummy regression data
import numpy as np
from sklearn.datasets import make_regression
X, y = make_regression(n_features=64, effective_rank =15,random_state=0, shuffle=False, n_samples=200)
```

### E3.1 compute PCA
see https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(X,  y)
```

```python
X_transformed = pca.transform(X)
X_transformed.shape
```

### E3.2 Plot Eigen Values

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(pca.explained_variance_)
```

### E3.3 select a $\lambda$ cut-off and project to new base

```python
cut_off = 30
pca_new = PCA(n_components = 30)
pca_new.fit(X, y)
```

```python
plt.plot(pca_new.explained_variance_ratio_)
```

```python
plt.plot(pca_new.singular_values_)
```
